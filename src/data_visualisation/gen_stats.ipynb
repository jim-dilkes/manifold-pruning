```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import csv
from matplotlib import cm
```

```python
data = []
for perc in range(0, 100, 10):
    with open("MFMA/Results/Masked/Testing4/masked_"+str(perc)+"_data_POS.csv", "r") as csvfile:
        reader = csv.reader(csvfile, delimiter=',')
        for indx, row in enumerate(reader):
            if indx == 0:
                continue
            data.append([perc]+row)
df_small = pd.DataFrame(data, columns=['Percentage', 'Layer', 'Capacity', 'Radius', 'Dimension', 'Correlation', 'NormedCapacity', 'NormedRadius', 'NormedDimension', 'NormedCorrelation'])
for col in ['Layer', 'Capacity', 'Radius', 'Dimension', 'Correlation', 'NormedCapacity', 'NormedRadius', 'NormedDimension', 'NormedCorrelation']:
    df_small[col] = df_small[col].astype(float)
```

```python
print(df_small.info())
```

```python
res_dir = "MFMA/Results/SizeMasked/"

folders = ["FinalTrain_1", "FinalTrain_2", "FinalTrain_3", "FinalTrain_4", "FinalTrain_5"]

single_plot = []
for folder_name in folders:
    tmp = []
    for layer in range(0,100,10):
        file = res_dir + folder_name + "/masked_" + str(layer) + "_data_POS.csv"
        with open(file, 'r') as csvfile:
            reader = csv.reader(csvfile, delimiter=',')
            for indx, row in enumerate(reader):
                if indx == 0:
                    continue
                tmp.append([folder_name, str(layer)] + row)
    single_plot.extend(tmp)
            

```

```python
df_test = pd.DataFrame(single_plot, columns = ['Mode', 'Percentage', 'Layer', 'Capacity', 'Radius', 'Dimension', 'Correlation', 'NormedCapacity', 'NormedRadius', 'NormedDimension', 'NormedCorrelation'])
```

```python
colors = cm.CMRmap(np.linspace(0,1,12))
```

```python
def plot(field, normed=False):
    n_runs = 5
    d_names = {'Capacity':3, 'Radius':4, 'Dimension':5, 'Correlation':6}
    val = d_names[field]
    if normed:
        val += 4
    for perc in range(0, 100, 10):
        line = df_test.loc[(df_test['Percentage'] == str(perc))].values[:, val].astype(np.float64).reshape(n_runs, 12).mean(axis=0)
        var = df_test.loc[(df_test['Percentage'] == str(perc))].values[:, val].astype(np.float64).reshape(n_runs, 12).var(axis=0)
        print(var)
        plt.plot(np.linspace(0,1,12), line, color=colors[int(perc//10)], label=str(perc) + '%')
        plt.fill_between(np.linspace(0,1,12), line-var, line+var, color=colors[int(perc//10)], alpha=0.2)
    plt.title(field + ' Plot')

    plt.legend(ncol=2, title='Pruned Percentage')
    if normed:
        plt.savefig('Results/FinalMasked-Size/Train/Normed'+field+'_plot.png')
    else:
        plt.savefig('Results/FinalMasked-Size/Train/'+field+'_plot.png')
    plt.close()
        
        
plot('Radius')
plot('Correlation')
plot('Dimension')
plot('Capacity')
plot('Radius', normed=True)
plot('Correlation', normed=True)
plot('Dimension', normed=True)
plot('Capacity', normed=True)
```
